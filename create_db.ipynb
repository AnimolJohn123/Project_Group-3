```python
import sqlite3
import pandas as pd
```

```python
def create_db(db_name,file_name,table_name):
    con = sqlite3.connect(db_name)
    cursor = con.cursor()
    carpredictions = pd.read_csv(file_name)
    carpredictions.to_sql(table_name,con,if_exists='replace', index=False)

    results = cursor.execute(f"select * from {table_name}").fetchall()
    con.commit()
    con.close()

    return results
```

```python
carpredictions= create_db("cars.db","car_predictions.csv","car_predictions")
```
